```python
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
print(maximum(10, 25, 15))
```

```python
def distinct_l(list):
    result=[]
    for item in list:
        if item not in result:
            result.append(item)
    return result
num=[23,45,67,23,54,12,67]
print(distinct_l(num))
```

```python
def mul_l(list):
    r=2
    for num in list:
        r=r*num
    return r
num=[20,30,40,50]
print(mul_l(num))
```

```python
def fac(n):
    r=1
    for i in range(1,n+1):
        r=r*i
    return r
num=int(input("enter a number:"))
print("factorial:",fac(num))
```

```python
def check_range(num):
    if 10 <= num <= 50:
        print(num, "falls in the range between 10 to 50.")
    else:
        print(num, "does not fall in the range between 10 to 50.")
check_range(45)        
```

```python
def print_even_num(num):
    for n in num:
        if n % 2 == 0:
            print(n)
my_lst = [10, 15, 20, 25, 30, 35, 40]
print("Even numbers are:")
print_even_num(my_lst)
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
def count_case_letters(text):
    upper_count = 0
    lower_count = 0
    for n in text:
        if n.isupper():
            upper_count += 1
        elif n.islower():
            lower_count += 1
    print("Uppercase letters:", upper_count)
    print("Lowercase letters:", lower_count)
count_case_letters("Python Programming")
```
