```python
name=input("enter your name:")
cla=int(input("enter your class:"))
sub=float(input("enter marks of subjects:"))
sub2=float(input("enter marks of subject2"))
sub3=float(input("enter marks of subject3"))
sub4=float(input("enter marks of subject4"))
sub5=float(input("enter marks of subject5"))
total=sub+sub2+sub3+sub4+sub5
percentage=total/500*100

if percentage>=60:
    grade ="A"
elif percentage>=50:
    grade="B"
elif percentage>=40:
    grade="C"
elif percentage>=33:
    grade="D"
else:
    grade="Fail"
print("\n-----result-----")
print("student_name:",name)
print("class:",cla)
print("total:",total)
print("percentage:",percentage)
print("grade:",grade)
```

```python
num = int(input("Enter a number: "))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
print("Reverse Number =", reverse)
if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```
