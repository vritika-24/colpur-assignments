```python
numbers = (10, 25, 5, 40, 15)
total = 0
highest = numbers[0]
lowest = numbers[0]
for num in numbers:
    total += num
    if num > highest:
        highest = num
    if num < lowest:
        lowest = num
print("Total :", total)
print("Highest Value :", highest)
print("Lowest Value :", lowest)
```

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)
new_result = ()
for i in n:
    if i > 10:
        new_result += (i,)
print("Filtered Tuple:",new_result)
```
