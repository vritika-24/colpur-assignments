```python
import requests
API_KEY = "YOUR_API_KEY"
symbol = input("Enter stock symbol (e.g., IBM, AAPL, MSFT): ").upper()
url = f"https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={symbol}&apikey={API_KEY}"
response = requests.get(url)
data = response.json()
if "Global Quote" in data and data["Global Quote"]:
    stock = data["Global Quote"]
    print("\n----- Live Stock Data -----")
    print("Symbol:", stock["01. symbol"])
    print("Open Price:", stock["02. open"])
    print("High Price:", stock["03. high"])
    print("Low Price:", stock["04. low"])
    print("Current Price:", stock["05. price"])
    print("Volume:", stock["06. volume"])
    print("Latest Trading Day:", stock["07. latest trading day"])
    print("Previous Close:", stock["08. previous close"])
    print("Change:", stock["09. change"])
    print("Change Percent:", stock["10. change percent"])
else:
    print("Invalid stock symbol or API limit exceeded.")
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    def to_fahrenheit(self):
        return (self.celsius * 9 / 5) + 32
    def to_kelvin(self):
        return self.celsius + 273.15
try:
    celsius = float(input("Enter temperature in Celsius: "))
    temp = Temperature(celsius)
    print("Temperature in Fahrenheit:", temp.to_fahrenheit())
    print("Temperature in Kelvin:", temp.to_kelvin())
except ValueError:
    print("Error: Please enter a valid numeric temperature.")
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
class Interest:
    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate
        self.time = time
    def simple_interest(self):
        s = (self.principal * self.rate * self.time) / 100
        return s
    def compound_interest(self):
        amount = self.principal * (1 + self.rate / 100) ** self.time
        c = amount - self.principal
        return c
try:
    p = float(input("Enter Principal Amount: "))
    r = float(input("Enter Rate of Interest (%): "))
    t = float(input("Enter Time (years): "))
    if p < 0 or r < 0 or t < 0:
        raise ValueError("Principal, Rate, and Time cannot be negative.")
    obj = Interest(p, r, t)
    print("Simple Interest =", obj.simple_interest())
    print("Compound Interest =", obj.compound_interest())
except ValueError as e:
    print("Error:", e)
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
class CoffeeMachine:
    def __init__(self, water, coffee, milk):
        self.water = water
        self.coffee = coffee
        self.milk = milk
    def make_latte(self):
        try:
            if self.water < 0 or self.coffee < 0 or self.milk < 0:
                raise ValueError("Resources cannot be negative.")
            if self.water >= 200 and self.coffee >= 20 and self.milk >= 150:
                self.water -= 200
                self.coffee -= 20
                self.milk -= 150
                print("Latte is ready!")
                print("Remaining Resources:")
                print("Water:", self.water, "ml")
                print("Coffee:", self.coffee, "g")
                print("Milk:", self.milk, "ml")
            else:
                raise Exception("Insufficient resources to make a latte.")
        except ValueError as e:
            print("Error:", e)
        except Exception as e:
            print("Error:", e)
try:
    water = int(input("Enter water (ml): "))
    coffee = int(input("Enter coffee (g): "))
    milk = int(input("Enter milk (ml): "))
    machine = CoffeeMachine(water, coffee, milk)
    machine.make_latte()
except ValueError:
    print("Error: Please enter valid integer values.")
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
class Vehicle:
    def __init__(self, name, max_speed):
        self.name = name
        self.max_speed = max_speed

    def display(self):
        print("Vehicle Name:", self.name)
        print("Maximum Speed:", self.max_speed, "km/h")
class Bus(Vehicle):
    def __init__(self, name, max_speed, seating_capacity):
        super().__init__(name, max_speed)
        self.seating_capacity = seating_capacity
    def reservation(self):
        print("Reservation Successful!")
        print("Seating Capacity:", self.seating_capacity)
try:
    name = input("Enter Vehicle Name: ")
    max_speed = int(input("Enter Maximum Speed (km/h): "))
    seating_capacity = int(input("Enter Seating Capacity: "))
    if max_speed <= 0 or seating_capacity <= 0:
        raise ValueError("Speed and Seating Capacity must be greater than 0.")
    bus = Bus(name, max_speed, seating_capacity)
    print("\nVehicle Details")
    bus.display()
    print("\nReservation Details")
    bus.reservation()
except ValueError as e:
    print("Error:", e)
except Exception as e:
    print("An unexpected error occurred:", e)
```
